```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

```python
cols = ['Brand', 'Selling Platform', 'Rating',
       'Refresh Rate (Hz)', 'Screen Size (inches)', 'RAM (GB)', 'Storage (GB)',
       'Processor']
```

```python
# COUNT PLOT

for features in cols:
    if features in df.columns:
        plt.figure(figsize=(15,5))
        sns.countplot(x=features,data=df)
        plt.title(f'Count plot of {features}')
        plt.xlabel(features)
        plt.ylabel("Count")
        plt.tight_layout()
        plt.show()




#HIST PLOT
feat = ['Brand', 'Price (USD)', 'Selling Platform', 'Rating',
       'Refresh Rate (Hz)', 'Screen Size (inches)', 'RAM (GB)', 'Storage (GB)',
       'Processor', 'Camera Setup']
for features in feat:
    if features in df.columns:
        plt.figure(figsize=(40,10))
        sns.histplot(df[features].dropna(),kde=True,color = 'green')
        plt.title(f'line plot of {features}')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()



# HEAT MAP

numeric_df = df.select_dtypes(include=[np.number])
if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(15,5))
    correlation_matrix = numeric_df.corr()
    sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
    plt.title("Correlation Heat map of Numeric Features")
    plt.tight_layout()
    plt.show()



#LINE PLOT

plt.figure(figsize=(15,5))
sns.lineplot(data=df, x='Model',y= 'Price (USD)',marker='o')
plt.show()
```
